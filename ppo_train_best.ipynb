```python
import gym

from stable_baselines3 import PPO
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.callbacks import EvalCallback
from stable_baselines3.common.monitor import Monitor
```

```python
continuous = False
```

```python
train_env = make_vec_env(f'LunarLander{"Continuous" if continuous else ""}-v2', n_envs=16)
eval_env = make_vec_env(f'LunarLander{"Continuous" if continuous else ""}-v2', n_envs=4)
eval_callback = EvalCallback(
    eval_env,
    best_model_save_path=f'./logs/{"continuous" if continuous else "discrete"}/ppo_opt',
    log_path=f'./logs/{"continuous" if continuous else "discrete"}/ppo_opt',
    eval_freq=int(1e5 // 16),
    n_eval_episodes=10,
    verbose=1
)
```

```python
model = PPO.load(f'./logs/{"continuous" if continuous else "discrete"}/ppo_opt/best_model', env=train_env)
```

```python
model.learn(total_timesteps=int(1e6), callback=eval_callback)
```
