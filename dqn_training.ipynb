```python
import gym

from stable_baselines3 import DQN
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.callbacks import EvalCallback
from stable_baselines3.common.monitor import Monitor
```

```python
continuous = False
```

```python
train_env = make_vec_env(f'LunarLander{"Continuous" if continuous else ""}-v2', n_envs=16)
eval_env = make_vec_env(f'LunarLander{"Continuous" if continuous else ""}-v2', n_envs=4)
```

```python
model = DQN(
    "MlpPolicy",
    train_env,
    learning_rate=6.3e-4,
    batch_size=128,
    buffer_size=50000,
    gamma=0.99,
    learning_starts=0,
    target_update_interval=250,
    train_freq=4,
    gradient_steps=-1,
    exploration_fraction=0.12,
    exploration_final_eps=0.1,
    policy_kwargs=dict(net_arch=[256, 256]),
    verbose=1,
    tensorboard_log=f'./runs/dqn_lunar_tensorboard/'
)
```

```python
eval_callback = EvalCallback(
    eval_env,
    best_model_save_path=f'./logs/discrete/dqn',
    log_path=f'./logs/discrete/dqn',
    eval_freq=int(1e5 // 16),
    n_eval_episodes=10,
    verbose=1
)
```

```python
model.learn(total_timesteps=int(1e6), callback=eval_callback)
```
